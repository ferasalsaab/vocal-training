```python
# Construct some test data
x, y = np.ogrid[-np.pi:np.pi:100j, -np.pi:np.pi:100j]
r = np.sin(np.exp((np.sin(x)**2 + np.cos(y)**3)))

# Find contours at a constant value of 0.8
contours = measure.find_contours(r, 0.8)

# Display the image and plot all contours found
fig, ax = plt.subplots()
ax.imshow(r, interpolation='nearest', cmap=plt.cm.gray)

for n, contour in enumerate(contours):
    ax.plot(contour[:, 1], contour[:, 0], linewidth=2)

ax.axis('image')
ax.set_xticks([])
ax.set_yticks([])
plt.show()
```

```python
import os
root_folder = 'sounds/recorded/large/'
total_count = 0
for folder in (folder for folder in os.listdir(root_folder) 
               if os.path.isdir(os.path.join(root_folder,folder))
              and 'hawker' not in folder ):
    files = os.listdir(os.path.join(root_folder, folder))
    count = 0
    for fle in files:
        if fle.endswith('.wav'):
            count += 1
            total_count += 1
    print('wav files in %s = %s'%(folder,str(count)))
print('total files = %s'%total_count)
```
